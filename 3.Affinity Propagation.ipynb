```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import AffinityPropagation
data = np.random.rand(200, 2)
annual_income = data[:, 0]  
spending_score = data[:, 1]
dataset = pd.DataFrame(data)
x= dataset.values  
aff_prop = AffinityPropagation()
aff_prop.fit(x)
labels = aff_prop.labels_
cluster_centers = aff_prop.cluster_centers_
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis', s=50,marker='.')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], color='red', marker='x', s=200, label='Cluster Centers')
plt.title('Affinity Propagation Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
n_clusters = len(np.unique(labels))
print(f"Number of clusters: {n_clusters}")
```
